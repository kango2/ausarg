```python
import pandas as pd

# Load CSV
df = pd.read_csv("/g/data/te53/ka6418/tmp/SGID/filtered_data.csv")

# Ensure participantID and tubeID are strings
df["participantID"] = df["participantID"].astype(str)
df["tubeID"] = df["tubeID"].astype(str)

# Fill missing tubeID values with a placeholder (if necessary)


# Generate SGID by ranking tubeID within each participantID
df["SGID_Rank"] = df.groupby("participantID")["tubeID"].rank(method="dense").fillna(0).astype(int)
df["SGID"] = df["participantID"] + "." + df["SGID_Rank"].astype(str)

# Drop temporary rank column
df.drop(columns=["SGID_Rank"], inplace=True)

# Save updated CSV
df.to_csv("illumina_with_sgid.csv", index=False)

print("✅ SGID column added and saved to illumina_with_sgid.csv")
```

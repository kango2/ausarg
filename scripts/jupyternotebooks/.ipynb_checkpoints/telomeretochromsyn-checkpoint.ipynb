```python
import pandas as pd
```

```python
telodf = pd.read_csv("/g/data/xl04/genomeprojects/Pogona_vitticeps/analysis/Telomeres/POGVITdef.h1.corrected_YAHS_Telomeres.csv")
faidf = pd.read_csv("/g/data/xl04/genomeprojects/Pogona_vitticeps/fasta/yahs/POGVITdef.h1.corrected_YAHS.fasta.fai", sep='\t', usecols=[0, 1], names=['SeqName', 'SeqLen'], header=None)
faidf[['Tel5', 'Tel3', 'Tel5Len', 'Tel3Len', 'Trim5', 'Trim3', 'TelPerc']] = False
telodf = telodf[['Sequence_ID','Relative Start']]
```

```python
for index, row in telodf.iterrows():
    seq_name = row["Sequence_ID"]
    rel_start = row["Relative Start"]
    if 0 <= rel_start <= 0.1:
        faidf.loc[faidf["SeqName"] == seq_name, "Tel5"] = True
    elif 0.9 <= rel_start <= 1:
        faidf.loc[faidf["SeqName"] == seq_name, "Tel3"] = True
```

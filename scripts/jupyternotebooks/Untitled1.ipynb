```python
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV with no header
df = pd.read_csv("/g/data/te53/variantcall/analysis/ONT-sexcheck/ratio.txt", header=None, names=["sample", "mnumber"])

# Plot as scatter
plt.figure(figsize=(10, 5))
plt.scatter(df.index,df["mnumber"], color='blue')

plt.xticks(rotation=90)  # Rotate x-axis labels if sample names are long
plt.xlabel("Sample")
plt.ylabel("mnumber")
plt.title("Scatter Plot of Samples vs mnumber")
plt.tight_layout()
plt.show()
```
